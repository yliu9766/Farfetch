```python
#print('before encoding', session['user'])
user_enc = LabelEncoder()
session['user'] = user_enc.fit_transform(session['user_id'].values)
n_users = session['user'].nunique()
#print('after encoding', session['user'])

item_enc = LabelEncoder()
session['product'] = item_enc.fit_transform(session['product_id'].values)
n_prod = session['product'].nunique()

min_rating = 0.0
max_rating = 1.0

n_users, n_prod, min_rating, max_rating
```

```python
X = session[['user', 'product', 'product_price']].values
y = session['is_click'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=420)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

n_factors = 25
#X_train_array = [X_train[:, 0], X_train[:, 1], X_train[:, 2]]
X_train_array = [X_train[:, 0], X_train[:, 1]]
#X_test_array = [X_test[:, 0], X_test[:, 1], X_test[:, 2]]
X_test_array = [X_test[:, 0], X_test[:, 1]]
```

```python
import tensorflow
from keras.models import Model
from keras.layers import Input, Reshape, Dot, Activation, Add, Lambda
from keras.layers.embeddings import Embedding
from keras.optimizers import Adam
from keras.regularizers import l2

class EmbeddingLayer:
    def __init__(self, n_items, n_factors):
        self.n_items = n_items
        self.n_factors = n_factors
    
    def __call__(self, x):
        x = Embedding(self.n_items, self.n_factors, embeddings_initializer='he_normal',
                      embeddings_regularizer=l2(1e-6))(x)
        x = Reshape((self.n_factors,))(x)
        return x

def RecommenderV2(n_users, n_prod, n_factors):
    user = Input(shape=(1,))
    u = EmbeddingLayer(n_users, n_factors)(user)
    ub = EmbeddingLayer(n_users, 1)(user)
    
    product = Input(shape=(1,))
    m = EmbeddingLayer(n_prod, n_factors)(product)
    mb = EmbeddingLayer(n_prod, 1)(product)
    
    x = Dot(axes=1)([u, m])
    x = Add()([x, ub, mb])
    x = Activation('softplus')(x)
    model = Model(inputs=[user, product], outputs=x)
    opt = Adam(lr=0.001)
    model.compile(loss='mean_squared_error', optimizer=opt)
    return model

# def RecommenderV1(n_users, n_prod, n_factors):
#     user = Input(shape=(1,))
#     u = Embedding(n_users, n_factors, embeddings_initializer='he_normal',
#                   embeddings_regularizer=l2(1e-6))(user)
#     u = Reshape((n_factors,))(u)
    
#     product = Input(shape=(1,))
#     m = Embedding(n_prod, n_factors, embeddings_initializer='he_normal',
#                   embeddings_regularizer=l2(1e-6))(product)
#     m = Reshape((n_factors,))(m)

#     x = Dot(axes=1)([u, m])
#     print(x)
#     x = Activation('softplus')(x)
#     model = Model(inputs=[user, product], outputs=x)
#     opt = Adam(learning_rate=0.001)
#     model.compile(loss='mean_squared_error', optimizer=opt)
#     return model
```

```python
deepLearner = RecommenderV2(n_users, n_prod, n_factors)

deepLearner.summary()
```
